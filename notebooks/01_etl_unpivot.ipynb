```python
import pandas as pd

# 1. Load the raw sales data from your data_raw folder
# Hiring Manager's Note: This file is "wide" (1,900+ columns), which is bad for SQL.
df = pd.read_csv(r'C:\Users\LENOVO\Desktop\M5 Forecasting Data\data-raw\sales_train_validation.csv')

# 2. Perform the "Melt" (Unpivot)
# We keep categories as they are and collapse all 'd_' day columns into one 'date_id'
melted_df = pd.melt(df, 
                    id_vars=['id', 'item_id', 'dept_id', 'cat_id', 'store_id', 'state_id'], 
                    var_name='d', 
                    value_name='sales')

# 3. Save the clean, vertical version to your data_processed folder
melted_df.to_csv('../data_processed/sales_unpivoted.csv', index=False)

print("Execution Successful: Your SQL-ready file is in data_processed!")
```
